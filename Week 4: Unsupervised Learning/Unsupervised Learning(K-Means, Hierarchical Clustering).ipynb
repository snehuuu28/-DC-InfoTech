# Learn about clustering algorithms (K-Means, Hierarchical Clustering).

Clustering is an unsupervised learning technique used to group data points into clusters based on their similarity. Two of the most commonly used clustering algorithms are K-Means and Hierarchical Clustering.

# 1. K-Means Clustering

1) K-Means is a centroid-based clustering algorithm that partitions the dataset into ùêæ
2) K clusters, where each data point belongs to the cluster with the nearest mean (centroid).

# Key Steps:

1) Initialize ùêæ centroids.
2) Assign each data point to the nearest centroid.
3) Recalculate the centroids based on the assigned data points.
4) Repeat until centroids no longer change or a predefined number of iterations are reached.

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate synthetic dataset
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Initialize K-Means with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)

# Fit the model and predict the clusters
y_kmeans = kmeans.fit_predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids')
plt.title('K-Means Clustering')
plt.legend()
plt.show()
```

# 2. Hierarchical Clustering

Hierarchical Clustering can be either:

1. Agglomerative (Bottom-Up): Start with each data point as its own cluster and merge them step by step.
2. Divisive (Top-Down): Start with all points in one cluster and split them iteratively.
3. We‚Äôll focus on Agglomerative Clustering, the most common approach.

# Key Steps:

1) Compute the distance matrix.
2) Merge the closest clusters.
3) Repeat until all clusters are merged into a single cluster.

```python
# Import libraries
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Use the same synthetic dataset as before (X)

# Perform Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=4)

# Fit and predict the clusters
y_hc = agg_clustering.fit_predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_hc, cmap='plasma')
plt.title('Hierarchical Clustering (Agglomerative)')
plt.show()

# Plot Dendrogram
Z = linkage(X, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Dendrogram')
plt.show()
```

# Key Differences Between K-Means and Hierarchical Clustering:

K-Means:
1) Requires the number of clusters to be specified beforehand.
2) Computationally efficient with large datasets.
3) Produces spherical clusters.
    
Hierarchical Clustering:

1) Does not require specifying the number of clusters in advance (but you can specify it).
2) Suitable for small to medium datasets.
3) Can produce more complex clusters (not necessarily spherical).

Both methods are widely used depending on the dataset and the desired cluster shapes.
